Ejercicio 2: Desarrolle una gráfica de linea usando el módulo pyplot de Matplotlib para producir la función:

$$ f(x) = x^2 |sen(x)| - 4 $$

Sobre el dominio $ 0 \leq x \leq 4 $. Asegurese de que los ejes y la curva esté correctamente etiquetadas. Agregue un titulo. 

```python
#Armar funcion f(x)

x = np.linspace(0,4,500)

f = x**2 * np.abs(np.sin(x))-4

plt.plot(x, f, label='f(x) = $x^2 * |sin(x)| - 4$', color='blue')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafico de funcion f(x)')
plt.grid(True)
plt.legend()

#Show 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos experimentales de la tabla
datos = np.array([
    [0, 1.3330],
    [5, 1.3403],
    [10, 1.3479],
    [15, 1.3557],
    [20, 1.3639],
    [25, 1.3723],
    [30, 1.3811],
    [35, 1.3902],
    [40, 1.3997],
    [45, 1.4096],
    [50, 1.4200],
    [55, 1.4307],
    [60, 1.4418],
    [65, 1.4532],
    [70, 1.4651],
    [75, 1.4774],
    [80, 1.4901],
    [85, 1.5033]
])

# Separar columnas
pctS = datos[:, 0]
n_medido = datos[:, 1]

# Modelo propuesto
pctS_modelo = np.linspace(0, 85, 200)
n_modelo = 1.324595 + 1.998515e-3 * pctS_modelo

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(pctS_modelo, n_modelo, label='Modelo propuesto', color='blue', linestyle='-')
plt.scatter(pctS, n_medido, label='Datos experimentales', color='red', marker='o')

# Etiquetas
plt.xlabel('Concentración de Sacarosa (%)')
plt.ylabel('Índice de Refracción (n)')
plt.title('Índice de Refracción vs. Concentración de Sacarosa')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Separar columnas
pctS = datos[:, 0]
n_medido = datos[:, 1]

# Modelo teórico
pctS_modelo = np.linspace(0, 85, 200)
n_modelo = 1.324595 + 1.998515e-3 * pctS_modelo

# Ajuste polinómico (grado 1: lineal)
coef = np.polyfit(pctS, n_medido, 1)
n_polyfit = np.polyval(coef, pctS_modelo)

# Gráfico
plt.figure(figsize=(9, 5))
plt.scatter(pctS, n_medido, color='red', label='Datos experimentales', zorder=3)
plt.plot(pctS_modelo, n_modelo, color='blue', linestyle='-', label='Modelo teórico')
plt.plot(pctS_modelo, n_polyfit, color='green', linestyle='--', label='Ajuste lineal (polyfit)')

# Etiquetas y estilo
plt.xlabel('Concentración de Sacarosa (%)')
plt.ylabel('Índice de Refracción (n)')
plt.title('Comparación entre Modelo Teórico y Ajuste Lineal')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Mostrar ecuación de ajuste
print(f"Ajuste lineal (polyfit): n = {coef[0]:.6f} * pctS + {coef[1]:.6f}")
```

Ejercicio 5:

Cree un gráfico en estructura de dos por dos usando **figure** para las siguientes funciones hiperbólicas en el dominio $ -1 \leq x \leq 1$

| Estructura del grafico | |
| ---------------------  | -- |
| $senh(x) \; vs \; x$ | $cosh(x) \; vs \; x$ |
| $tanh(x) \; vs \; x$ | $sech(x) \; vs \; x$ |

```python
import numpy as np
import matplotlib.pyplot as plt

# Dominio
x = np.linspace(-1, 1, 400)

# Funciones hiperbólicas
sinh_x = np.sinh(x)
cosh_x = np.cosh(x)
tanh_x = np.tanh(x)
sech_x = 1 / np.cosh(x)

# Crear figura y subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Funciones Hiperbólicas', fontsize=14)

# Gráfica sinh(x)
axs[0, 0].plot(x, sinh_x, color='blue')
axs[0, 0].set_title('sinh(x) vs x')
axs[0, 0].grid(True)

# Gráfica cosh(x)
axs[0, 1].plot(x, cosh_x, color='orange')
axs[0, 1].set_title('cosh(x) vs x')
axs[0, 1].grid(True)

# Gráfica tanh(x)
axs[1, 0].plot(x, tanh_x, color='green')
axs[1, 0].set_title('tanh(x) vs x')
axs[1, 0].grid(True)

# Gráfica sech(x)
axs[1, 1].plot(x, sech_x, color='red')
axs[1, 1].set_title('sech(x) vs x')
axs[1, 1].grid(True)

# Etiquetas generales
for ax in axs.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
